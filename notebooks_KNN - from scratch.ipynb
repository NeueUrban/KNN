generate synthetic data

```python
mu = np.array([0,0])
```

```python
cov = np.array([[1,0],[0,10] ])
```

```python
np.linalg.cholesky(cov)
```

```python
data = np.random.multivariate_normal(mu,cov,size=1000)
```

```python
data
```

```python
x1 = data[:,0]
```

```python
x2 = data[:,1]
```

```python
plt.scatter(x1,x2)
plt.axis("equal")
```
